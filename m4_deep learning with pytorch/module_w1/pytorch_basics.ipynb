### Derivatives

```python
# compute derivative of a single variable
x = torch.tensor(6.0, requires_grad=True)
y = x**2 - 1
y.backward()
x.grad
```

```python
x = torch.tensor(5.0, requires_grad=True)
y = torch.tensor(-3.0, requires_grad=True)
z = x**2 + y**2 - x*y

z.backward()
print(f"x derivative: {x.grad}")
print(f"y derivative: {y.grad}")
```

```python
# constants for evaluating derivative
x1 = torch.tensor(3.5, requires_grad=True)
x2 = torch.tensor(2.0, requires_grad=True)
x3 = torch.tensor(4.5, requires_grad=True)

# function in terms of x1, x2, x3
H = x1**2 +x2**2 +x3**2 + 3*x1*x2*x3
H.backward()

# evaluate the derivatives respected to x1, x2 and x3
print(f"Derivative x1: {x1.grad}")
print(f"Derivative x2: {x2.grad}")
print(f"Derivative x3: {x3.grad}")
```
